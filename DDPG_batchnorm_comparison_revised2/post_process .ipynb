```python
import matplotlib.pyplot as plt 
import pickle 
import numpy as np 
```

```python
with open("DDPG1_modelChange.pkl", "rb") as f: 
    data_w = pickle.load(f)

with open("DDPG2_modelChange.pkl", "rb") as f: 
    data_kw = pickle.load(f)

with open("DDPG3_modelChange.pkl", "rb") as f: 
    data_norm = pickle.load(f)
    
with open("DDPG4_modelChange.pkl", "rb") as f: 
    data_batchnorm = pickle.load(f)
```

# W VS kW 

```python
plt.figure(figsize=(12, 8))

plt.plot(data_w[2]["rewards"], linewidth=3, label="Power in [W]")
plt.plot(data_kw[2]["rewards"], linewidth=3, label="Power in [kW]")

plt.xlabel("Episodes", fontsize=16)
plt.ylabel("Episodic rewards", fontsize=16)
plt.legend(fontsize=16)
plt.xlim([0, 120])
plt.grid()
plt.savefig("results_W_vs_kW.png")
```
