```python
import matplotlib.pyplot as plt 
import pickle 
import matplotlib as mpl 
import numpy as np 
```

```python
with open("DDPG_cycleOne_1to3.pkl", "rb") as f: 
    data1to3 = pickle.load(f)
with open("DDPG_cycleOne_4to6.pkl", "rb") as f: 
    data4to6 = pickle.load(f)
with open("DDPG_cycleOne_7to10.pkl", "rb") as f: 
    data7to10 = pickle.load(f)

data = {} 
for data_temp in [data1to3, data4to6, data7to10]: 
    for k, v in data_temp.items(): 
        data[k] = v 

    
```
